
# Simple Linear Regression – Marketing ROI Analysis

## Project Overview
This project analyzes a marketing dataset using Python and `statsmodels` to build a Simple Linear Regression model. The goal is to identify which marketing channel has the strongest relationship with Sales and provide ROI-based recommendations for budget allocation.

### Objectives
- Load and clean the dataset
- Perform exploratory data analysis (EDA)
- Identify the most correlated marketing variable
- Build an OLS regression model
- Validate regression assumptions
- Interpret statistical outputs
- Provide business recommendations


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy import stats

# Load dataset
df = pd.read_csv('85334965-5736-457a-b8d4-a077e6872f84.csv')

# Display first rows
df.head()
```

```

# Dataset information
print(df.info())

# Check missing values
print(df.isnull().sum())

# Drop missing values if any
df = df.dropna()

# Summary statistics
df.describe()
```


# Exploratory Data Analysis (EDA)

We visualize the relationship between marketing channels and Sales to identify the strongest predictor.


```

# Correlation matrix
corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()

# Scatterplots
fig, axes = plt.subplots(1, 3, figsize=(18,5))

sns.scatterplot(x='TV', y='Sales', data=df, ax=axes[0])
axes[0].set_title('TV vs Sales')

sns.scatterplot(x='Radio', y='Sales', data=df, ax=axes[1])
axes[1].set_title('Radio vs Sales')

sns.scatterplot(x='Social_Media', y='Sales', data=df, ax=axes[2])
axes[2].set_title('Social Media vs Sales')

plt.tight_layout()
plt.show()

# Identify strongest correlation
print(corr['Sales'].sort_values(ascending=False))
```


# Building the Simple Linear Regression Model

Based on the correlation analysis, we select the independent variable with the strongest correlation to Sales.


```

# Selecting TV as independent variable
X = df[['TV']]
y = df['Sales']

# Add constant
X = sm.add_constant(X)

# Fit OLS model
model = sm.OLS(y, X).fit()

# Model summary
print(model.summary())
```


# Diagnostic Plots

The following diagnostic plots help validate regression assumptions:
- Linearity
- Normality of residuals
- Homoscedasticity


```

# Predictions and residuals
predictions = model.predict(X)
residuals = y - predictions

# Residual plot
plt.figure(figsize=(6,4))
sns.scatterplot(x=predictions, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# Histogram of residuals
plt.figure(figsize=(6,4))
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.show()

# QQ plot
sm.qqplot(residuals, line='45')
plt.title('QQ Plot')
plt.show()

# Breusch-Pagan test for homoscedasticity
bp_test = het_breuschpagan(residuals, X)

labels = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
print(dict(zip(labels, bp_test)))
```


# Interpretation of Results

## R-squared
The R-squared value explains how much of the variation in Sales is explained by the selected marketing channel.

## Coefficient
The regression coefficient shows the expected increase in Sales for every one-unit increase in the marketing budget.

## p-value
A p-value below 0.05 indicates that the relationship between the marketing channel and Sales is statistically significant.

## Business Recommendation
If TV advertising has the highest correlation and significant coefficient, the company should prioritize TV marketing investments because it provides the strongest measurable impact on Sales.


```

# Final conclusion output
print("Marketing ROI Analysis Complete.")
print("Recommended marketing channel:", corr['Sales'].drop('Sales').idxmax())
```
